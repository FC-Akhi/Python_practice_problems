## 13. What's the difference between tuples and lists?

**Ans:**

**Mutability:** Means elements can be modified, added, or removed after creation.
- Lists are mutable
- Tuples are immutable

**Syntax**
- Lists are defined using square brackets []
- Tuples are defined using parentheses ()


```python
# declaring a list
my_list = [1, 2, 3]
my_list[0] = 4  # modifying an element
my_list.append(5)  # adding an element
my_list.remove(2)  # removing an element
print(my_list)  # output: [4, 3, 5]
print(type(my_list))

# declaring a tuple
my_tuple = (1, 2, 3)
# my_tuple[0] = 4  # trying to modify an element will raise an error
# my_tuple.append(4)  # tuples do not have an append() method
# my_tuple.remove(2)  # tuples do not have a remove() method
print(my_tuple)  # output: (1, 2, 3)
print(type(my_tuple))
```

## 14. How do you type a tuple value that only contains the integer 42?

## 15. How do you get a list value's tuple form? How do you get a tuple value's list form?

```python
# typecasting list to tuple
my_list = [1, 2, 3]
my_tuple = tuple(my_list)
print(my_tuple)  # output: (1, 2, 3)

# typecasting tuple to list
my_tuple = (1, 2, 3)
my_list = list(my_tuple)
print(my_list) # output: [1, 2, 3]
```

## 16. Variables that "contain" list values are not necessarily lists themselves. Instead, what do they contain?

**Ans:** Variables that contain list values in Python actually **store references** to the list objects rather than the list objects themselves. In Python, variables are essentially names or labels that refer to objects in memory.

When you assign a list to a variable, the variable holds a reference to the memory location where the list object is stored. This means that the variable does not directly contain the list values but rather points to the memory location where the list values are stored.

**Note:** The id() function can be used to retrieve the memory address of an object, but it does not provide a way to obtain the memory address of the variable that references the object.



```python
# declare a list
my_list = [1, 2, 3]

print(id(my_list))
print(id(my_list[0]))
print(id(my_list[1]))
```

## 17. How do you distinguish between copy.copy() and copy.deepcopy()?

**Ans:** The copy.copy() and copy.deepcopy() functions from the copy module in Python are used to create copies of objects.

```python
import copy

# copy.copy()
original_list = [1, 2, [3, 4]]
copied_list = copy.copy(original_list)

original_list[0] = 5
original_list[2].append(6)


print(original_list)
print(copied_list)  # Output: [1, 2, [3, 4, 6]]
```

```python
# copy.deepcopy()
original_list = [1, 2, [3, 4]]
copied_list = copy.deepcopy(original_list)

original_list[0] = 5
original_list[2].append(6)


print(original_list)
print(copied_list)
```
