## 5. How many global scopes are there in a Python program? How many local scopes?

**Ans: Global scope:** There is usually **one** global scope in a Python program. Variables and functions defined in the global scope can be accessed from anywhere within the program.

**Local scopes:** There can be multiple local scopes in a Python program. Each time a function is called or a control flow structure is entered, a new local scope is created. Variables defined within a local scope are only accessible within that scope and its nested scopes.

```python
# global variable
s = 100

def func(s):
    # local variable
    a = 10
    print(a)
    print(s + a)


func(s)
print(s)
```

## 7. What is the concept of a return value? Is it possible to have a return value in an expression?
**Ans:** The concept of a return value refers to the value that a function can optionally produce and provide back to the caller. When a function is called, it may perform some computations, manipulate data, or perform other operations, and then return a result.

We can have result of an expression as return value. But, we can't return a whole expression.


```python
def sum(a, b):
    
    # sum result will return via expression a + b. Here, a and b is operand and + is operator
    return a + b

# function call and value returned will be stored in result
result = sum(3, 2)
print(result)
```

## 9. How do you make a function variable refer to the global variable?


```python
# define global variable
global_var = 10

def my_function():
    
    # use global varaible as local variable
    global global_var
    
    # Access and modify the global variable
    global_var += 5
    print(global_var)

my_function()  # Output: 15

print(global_var) # Output: 15
```

**Ans:**

To prevent a program from crashing when it encounters an error, we can implement error handling techniques. By handling, we can allow the program to continue execution or take alternative actions instead of abruptly terminating.

For example:

**Try-Except blocks:** Surround the potentially error-prone code with a try-except block. The code inside the try block is executed, and if an exception occurs, it is caught by the except block. You can handle the exception by providing appropriate error handling logic within the except block.

**try:**

## 14. What is the purpose of the try clause? What is the purpose of the except clause?

**Ans:** try-except usually works together. 

In a simple words;

The purpose of the try clause is to enclose a block of code that might potentially raise an exception.
